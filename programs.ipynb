```python
# Nodes are courses, programs, departments, colleges, and sections
#There are edges from collegde to department, department to program, program to course, and course to section
# g2022f, g2023s: Links by requirements
# g_registrar: Links by courses taken

#Jaccard similarity
#This puts it in a graph so that we can use networkx to find the shortest path between the two programs. p is the similarity. 

g_programs = nx.subgraph_view(Gf2022, filter_node=lambda x: x.startswith("p") or x.startswith("c"))
programs = list(filter(lambda x: x.startswith("p"), g_programs.nodes))
ebunch = []
for i in range(len(programs)):
    for j in range(i + 1, len(programs)):
        ebunch.append((programs[i], programs[j]))

program_similarity = nx.Graph()
for u, v, p in nx.jaccard_coefficient(g_programs, ebunch):
    program_similarity.add_edge(u, v, weight=p)
    
```

```python
# Code:
# 1. Take a program
# 2. Get a list of courses required for that program
# 3. Get a list of programs that require those courses
# 4. Return that list of programs
# 5. Repeat for each program in the list of programs
# 6. Return a list of all programs that were returned by the function

#Algorithm: 
program = "p33681"
courses = list(g_programs.neighbors(program))
programs = []
for course in courses:
    programs += list(g_programs.neighbors(course))
    
programs = list(set(programs))
programs.remove(program)
programs
#Similarity between the programs
for program2 in programs:
    print(f"{g_programs.nodes[program]['attr']['desc']} and {g_programs.nodes[program2]['attr']['desc']} have a similarity of {program_similarity[program][program2]['weight']}")
    
#Shortest path between the programs
for program2 in programs:
    
    print(f"{g_programs.nodes[program]['attr']['desc']} and {g_programs.nodes[program2]['attr']['desc']} have a shortest path of {nx.shortest_path_length(program_similarity, program, program2, weight='weight')}")
    

```

```python
# Now I need to find the similarity between the program p33681 and a minor check this if progam_type is 'Minor'
program = "p33681"
minors = []
for minor in g_programs.nodes:
    if minor.startswith('p'):
        if g_programs.nodes[minor]['attr']['program_type'] == 'Minor':
            minors.append(minor)
        
program = "p33681"
if program in minors:
    minors.remove("p33681")
minors
#Show the similarity between the programs
for minor2 in minors:
    
    print(f"{g_programs.nodes[program]['attr']['desc']} and {g_programs.nodes[minor2]['attr']['desc']} have a similarity of {program_similarity[program][minor2]['weight']}")
    
```
